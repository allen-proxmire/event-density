# ED-SIM-02 Visualization Gallery

This notebook demonstrates the visualization layer of ED-SIM-02:
- Field snapshots (initial and final)
- Invariant time-evolution plots (energy, complexity, entropy)
- Morphology fraction evolution
- Dissipation channel decomposition
- Correlation and structure function diagnostics
- Animated field and dashboard evolution

All plots are generated from a single simulation run using the
`A_2d_cosine` scenario.

```python
import numpy as np
import matplotlib.pyplot as plt

from edsim.experiments.scenarios import get_scenario
from edsim.experiments.atlas import run_atlas, summarize_time_series
from edsim.visualization import (
    plot_field_snapshot,
    plot_field_timeseries,
    plot_field_comparison,
    plot_energy,
    plot_complexity,
    plot_spectral_entropy,
    plot_dissipation_channels,
    plot_correlation_length,
    plot_structure_function,
    plot_atlas_dashboard,
    plot_morphology_fractions,
    animate_field,
    animate_morphology,
    animate_invariant_dashboard,
)

%matplotlib inline
```

```python
# Run the canonical 2D scenario
scenario = get_scenario("A_2d_cosine")
params, ts = run_atlas(scenario)
print(f"Scenario: {scenario.name}")
print(f"d={params.d}, N={params.N}, T={params.T}, {len(ts.times)} snapshots")
```

## Field Snapshots

The density field at t=0, t=T/2, and t=T, showing relaxation toward
the equilibrium rho_star = 0.5.

```python
fig = plot_field_comparison(ts, steps=[0, len(ts.times)//2, -1])
plt.show()
```

## Core Invariants

Energy and complexity decrease monotonically (Law 2). Spectral entropy
decreases as high-k modes decay faster than low-k (Law 3).

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
plot_energy(ts, ax=axes[0])
plot_complexity(ts, ax=axes[1])
plot_spectral_entropy(ts, ax=axes[2])
plt.tight_layout()
plt.show()
```

## Morphology Evolution

In 2D, the field is classified into blobs (local maxima with both
Hessian eigenvalues negative) and sheets (saddle regions).

```python
fig = plot_morphology_fractions(ts)
plt.show()
```

## Dissipation Channels and Correlation Length

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
plot_dissipation_channels(ts, ax=axes[0])
plot_correlation_length(ts, ax=axes[1])
plt.tight_layout()
plt.show()
```

## Structure Function

S_2(r) at the initial and final times. The structure function captures
the scale-dependent variance of the density field.

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
plot_structure_function(ts, step=0, ax=axes[0])
plot_structure_function(ts, step=-1, ax=axes[1])
plt.tight_layout()
plt.show()
```

## Full Atlas Dashboard

All six core diagnostics in a single figure.

```python
fig = plot_atlas_dashboard(ts)
plt.show()
```

## Animations

Animated density field evolution. In Jupyter, display with
`HTML(ani.to_jshtml())` or save with `ani.save("field.gif")`.

```python
ani = animate_field(ts, interval=150)
# To display in Jupyter:
from IPython.display import HTML
HTML(ani.to_jshtml())
```

## Summary

This gallery demonstrates the visualization primitives of ED-SIM-02.
Every function accepts a `TimeSeries` object and returns a matplotlib
`Figure`, making it easy to compose custom layouts or save figures
for publication.

```python
summary = summarize_time_series(ts)
for k in ['energy_ratio', 'complexity_ratio', 'R_grad_mean', 'xi_ratio', 'energy_monotone']:
    print(f"  {k}: {summary[k]}")
```
